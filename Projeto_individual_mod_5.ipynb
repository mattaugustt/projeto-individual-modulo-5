Gráfico 1: <br>
Podemos ver que ele relaciona o impacto que o motivo de os funcionários saírem da empresa causaria e a quantidade de pessoas que saíram por cada um dos motivos, segundo o Parceiro de Negócios de Recursos Humanos. <br>
Podemos ver que assuntos mais naturais são taxados com um pequeno impacto (como doenças, trajeto até o trabalho, mudança de carreira etc) sobre a empresa e que o parceiro de negócios identificou que uma pequena quantidade de pessoas deixa a empresa por esses motivos. <br>
Já os assuntos mais relacionados a carreira são os que mais representam risco para a empresa perder este trabalhador, como por exemplo o treinamento para tal função (visto que o profissional pode achar que seja algo que exija demais de si próprio, ou que não ache que tenha competência para chegar ao fim), a falta de reconhecimento também é um dos grandes motivos (com a falta de valorização dos trabalhos prestados pelo funcionário), além desses, o conflito com outras pessoas no ambiente de trabalho também é um grande risco para a empresa perder este funcionário, porém na visão do parceiro de negócios os funcionários que saíram por esses motivos foram poucos. <br>
Na visão do Parceiro de Negócios de Recursos Humanos, os motivos que levou uma grande quantidade de funcionários a saírem da empresa foram: o avanço na carreira, o tipo de trabalho exercido e a realocação no ambiente de trabalho, que apesar de terem menores chances de impactar na empresa, acabam impactando por ter uma grande debandada de funcionários por estes motivos. <br><br>
Esses dados não são tão simples de serem explicados pois tem o fato de haver o aumento dos círculos com o aumento do eixo x (HRBP), que poderia ser interpretado apenas pela posição em relação ao eixo. Isso pode causar algum tipo de interpretação errada sobre os dados, visto que ambos os dados dos dois eixos são importantes.

Gráfico 2: <br>
Vemos que nesse gráfico é feita a relação entre o impacto que cada motivo do funcionário causaria na empresa e a quantidade de funcionários que saiu por cada um dos motivos (possivelmente, somente dos funcionários que decidiram comunicar o motivo). <br>
Identificamos que a maior parte dos funcionários que saíram da empresa assumiram que decidiram sair da mesma por terem avançado em suas carreiras e uma outra parcela razoavelmente grande alegou que não estava satisfeita com o tipo de trabalho que exerciam na mesma. <br>
Os dois motivos com maior risco de ocorrerem são uns dos que possuem menos trabalhadores tendo saído por eles, enquanto os motivos com risco mediano são os que apresentam maior quantidade de empregados saindo por conta deles.




###Que observações você pode fazer a partir desses dados? Coloque três conclusões específicas em palavras (Escreva sentenças completas).


<br>
Podemos concluir que, na maioria das vezes, tanto na visão do parceiro de negócios de recursos humanos quanto na visão da pesquisa feita, a razão do funcionário sair da empresa é por motivos de carreira. <br>
Além disso, vemos que:  <br>
- O risco não está estritamente relacionado com a quantidade de pessoas que de fato saíram ou sairão de uma empresa, são situações imprevisíveis e que não podem ser bem rotuladas e classificadas. <br>
- Já os dados de motivos dos funcionários decidirem deixar a empresa podem ser utilizados como forma de tentar ver o que pode ser mudado (fatores que não sejam externos da empresa). <br>
- Profissionais que não se sentem valorizados pela empresa, por não serem remunerados de forma justa ou não serem promovidos apesar de competentes, ou não se sentem confortáveis com as exigências da empresa, seja em questão de tempo exigido ou de competências exigidas, costumam ter maior taxa de evasão.

###O objetivo é pensar além de mostrar simplesmente dados. Como você pode gerar essas informações em uma história orientada por dados para seu gerente, o CMO. (Qual é o contexto da história?).  <br>
<br>
Nossa empresa deseja descobrir o motivo de tantos funcionários estarem decidindo se desligarem da empresa, e por isso resolvemos fazer uma pesquisa com os que saíram para saber seus motivos e, em caso de insatisfação, tentar mudar as características de nossa instituição para que ela seja mais atrativa ao mercado e mais agradável aos nossos colaboradores. <br>
O responsável pelo setor de Recursos Humanos (RH) fez uma análise das situações de cada um dos que foram desligados para tentar descobrir qual era o possível motivo que os fizeram tomar tal decisão. <br>
Após isso, analisamos a pesquisa que fizemos com cada um dos ex-funcionários e a pesquisa comportamental que o parceiro de negócios de recursos humanos levantou. <br>
Com esses dados, nós obtivemos esses gráficos que mostram quais são os motivos que a pesquisa comportamental e a pesquisa pessoal deram como causadores da evasão em massa dos funcionários. <br>
Em ambos os casos, vimos que o avanço na carreira e o tipo de trabalho que o funcionário executava eram os maiores causadores dessa evasão. Essas duas situações podem ser contornadas caso tenhamos a possibilidade de promover os funcionários que estão aptos para uma atividade mais importante e também tentar reformular a atividade de trabalho, ou seja, a forma de produção, o ambiente de trabalho e tentar tornar o momento de descanso do funcionário mais confortável para que ele se sinta menos exausto.
